# Module 0.6: Exponentielles, Logarithmes et Trigonométrie
## Fonctions Essentielles pour le ML

Ce notebook illustre les fonctions exponentielles, logarithmes et fonctions trigonométriques.

```python
# Imports
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import FancyBboxPatch
import warnings
warnings.filterwarnings('ignore')

plt.style.use('seaborn-v0_8-darkgrid')
plt.rcParams['figure.figsize'] = (12, 6)
```

## 1. Fonction Exponentielle

```python
# Visualisation de e^x
x = np.linspace(-3, 3, 400)
# TODO: Calculer
y = np.exp(x)

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Plot 1: Fonction exponentielleaxes[0].plot(x, y, 'b-', linewidth=3, label='y = eˣ')
axes[0].axhline(y=1, color='r', linestyle='--', alpha=0.5, label='y = 1 (e⁰ = 1)')
axes[0].plot(x, y, 'b-', linewidth=3, label='y = eˣ')
axes[0].set_title('Fonction Exponentielle')
axes[0].set_xlabel('x')
axes[0].set_ylabel('eˣ')
axes[0].legend()
axes[0].grid(True)

# ... (Le reste du code de visualisation est laissé pour aider l'étudiant)
```

## 2. Fonction Logarithme

```python
# Visualisation du logarithme népérien
x = np.linspace(0.01, 5, 400)

# TODO: Calculer y_ln = ln(x)
y_ln = np.log(x)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Plot 1: ln(x)
axes[0].plot(x, y_ln, 'g-', linewidth=3, label='y = ln(x)')
axes[0].axhline(y=0, color='r', linestyle='--', alpha=0.5, label='y = 0 (ln(1) = 0)')
```

## 3. Application : Fonction Sigmoïde (Logistique)

```python
# La fonction sigmoïde est centrale en ML
def sigmoid(x):
    # TODO: Implémenter la fonction sigmoïde: 1 / (1 + exp(-x))
    return 1 / (1 + np.exp(-x))

x = np.linspace(-10, 10, 400)
y_sigmoid = sigmoid(x)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Plot 1: Sigmoïde
axes[0].plot(x, y_sigmoid, 'b-', linewidth=3, label='σ(x) = 1/(1+e⁻ˣ)')
axes[0].plot(x, np.ones_like(x), 'r--', linewidth=2, label='y = 1')
# ... (Le reste du code de visualisation est laissé pour aider l'étudiant)
```

## 4. Fonctions Trigonométriques

```python
# Visualisation de sin, cos, tan
x = np.linspace(-2*np.pi, 2*np.pi, 400)
# TODO: Calculer y_sin = sin(x)
y_sin = lambda x: np.sin(x)

# TODO: Calculer y_cos = cos(x)
y_cos = lambda x: np.cos(x)
# Calculer y_tan = tan(x)
y_tan = lambda x: np.tan(x)

fig, axes = plt.subplots(2, 2, figsize=(14, 10))
# Plot 1: Cercle trigonométrique
theta = np.linspace(0, 2*np.pi, 100)
axes[0, 0].plot(np.cos(theta), np.sin(theta), 'b-', linewidth=2, label='Cercle trigonométrique')
axes[0, 0].set_aspect('equal')
axes[0, 0].set_xlim(-1.1, 1.1)
axes[0, 0].set_ylim(-1.1, 1.1)
axes[0, 0].set_xlabel('cos(θ)')
axes[0, 0].set_ylabel('sin(θ)')
axes[0, 0].set_title('Cercle Trigonométrique')
axes[0, 0].legend()

# Plot 2: sin(x)
axes[0, 1].plot(x, y_sin(x), 'b-', linewidth=3, label='sin(x)')
axes[0, 1].set_title('Fonction Sinus')
axes[0, 1].set_xlabel('x')
axes[0, 1].set_ylabel('sin(x)')
axes[0, 1].axhline(0, color='black',linewidth=0.5)
axes[0, 1].axvline(0, color='black',linewidth=0.5)
axes[0, 1].legend()
axes[0, 1].grid(True)

# Plot 3: cos(x)
axes[1, 0].plot(x, y_cos(x), 'g-', linewidth=3, label='cos(x)')
axes[1, 0].set_title('Fonction Cosinus')
axes[1, 0].set_xlabel('x')
axes[1, 0].set_ylabel('cos(x)')
axes[1, 0].axhline(0, color='black',linewidth=0.5)
axes[1, 0].axvline(0, color='black',linewidth=0.5)
axes[1, 0].legend()
axes[1, 0].grid(True)

# Plot 4: tan(x)
y_tan_val = y_tan(x)
y_tan_val[np.abs(y_tan_val) > 10] = np.nan # Set large values to NaN for cleaner plot
axes[1, 1].plot(x, y_tan_val, 'r-', linewidth=3, label='tan(x)')
axes[1, 1].set_title('Fonction Tangente')
axes[1, 1].set_xlabel('x')
axes[1, 1].set_ylabel('tan(x)')
axes[1, 1].axhline(0, color='black',linewidth=0.5)
axes[1, 1].axvline(0, color='black',linewidth=0.5)
axes[1, 1].set_ylim(-5, 5) # Set reasonable y-limits
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```

## 5. Identité de Pythagore

```python
# Vérification de sin²(x) + cos²(x) = 1
x = np.linspace(0, 4*np.pi, 400)
# TODO: Calculer sin(x)^2
sin_squared = np.sin(x)**2
# TODO: Calculer cos(x)^2
cos_squared = np.cos(x)**2
# TODO: Calculer la somme
sum_squares = sin_squared + cos_squared
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Plot 1: sin²(x) et cos²(x)
axes[0].plot(x, sin_squared, 'b-', linewidth=2, label='sin²(x)')
axes[0].plot(x, cos_squared, 'r-', linewidth=2, label='cos²(x)')
axes[0].set_title('sin²(x) et cos²(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('Valeur')
axes[0].legend()
axes[0].grid(True)

# Plot 2: sin²(x) + cos²(x)
axes[1].plot(x, sum_squares, 'g-', linewidth=3, label='sin²(x) + cos²(x)')
axes[1].axhline(1, color='black', linestyle='--', linewidth=1, label='Valeur de 1') # Add a line at y=1
axes[1].set_title('sin²(x) + cos²(x) = 1')
axes[1].set_xlabel('x')
axes[1].set_ylabel('Somme')
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## 6. Dérivées des Fonctions Trigonométriques

```python
# Visualisation des dérivées
x = np.linspace(-2*np.pi, 2*np.pi, 400)
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
# Plot 1: sin(x) et sa dérivée
# TODO: Tracer sin(x)
axes[0].plot(x, np.sin(x), 'b-', linewidth=2, label='sin(x)')
# TODO: Tracer la dérivée de sin(x) (c'est cos(x))
axes[0].plot(x, np.cos(x), 'r--', linewidth=2, label="(sin(x))' = cos(x)")
axes[0].set_title('Dérivée de sin(x)')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].axhline(0, color='black',linewidth=0.5)
axes[0].axvline(0, color='black',linewidth=0.5)
axes[0].legend()
axes[0].grid(True)

# Plot 2: cos(x) et sa dérivée
axes[1].plot(x, np.cos(x), 'b-', linewidth=2, label='cos(x)')
axes[1].plot(x, -np.sin(x), 'm--', linewidth=2, label="(cos(x))' = -sin(x)")
axes[1].set_title('Dérivée de cos(x)')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].axhline(0, color='black',linewidth=0.5)
axes[1].axvline(0, color='black',linewidth=0.5)
axes[1].legend()
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## 7. Application : Ondes et Signaux

```python
# Combinaison de sinusoïdes (série de Fourier simple)
t = np.linspace(0, 4*np.pi, 1000)# Signal composite
# TODO: Définir la fondamentale sin(t)
fundamental = np.sin(t)
# TODO: Définir la 2ème harmonique 0.5 * sin(2t)
harmonic2 = 0.5 * np.sin(2*t)
# TODO: Définir la 3ème harmonique 0.3 * sin(3t)
harmonic3 = 0.3 * np.sin(3*t)
# TODO: Somme des signaux
composite = fundamental + harmonic2 + harmonic3
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
# Composantes
axes[0, 0].plot(t, fundamental, label='Fondamentale: sin(t)', color='blue')
axes[0, 0].set_title('Fondamentale (sin(t))')
axes[0, 0].set_xlabel('Temps (t)')
axes[0, 0].set_ylabel('Amplitude')
axes[0, 0].legend()
axes[0, 0].grid(True)

axes[0, 1].plot(t, harmonic2, label='2ème Harmonique: 0.5 * sin(2t)', color='green')
axes[0, 1].set_title('2ème Harmonique (0.5 * sin(2t))')
axes[0, 1].set_xlabel('Temps (t)')
axes[0, 1].set_ylabel('Amplitude')
axes[0, 1].legend()
axes[0, 1].grid(True)

axes[1, 0].plot(t, harmonic3, label='3ème Harmonique: 0.3 * sin(3t)', color='red')
axes[1, 0].set_title('3ème Harmonique (0.3 * sin(3t))')
axes[1, 0].set_xlabel('Temps (t)')
axes[1, 0].set_ylabel('Amplitude')
axes[1, 0].legend()
axes[1, 0].grid(True)

axes[1, 1].plot(t, composite, label='Signal Composite', color='purple', linewidth=2)
axes[1, 1].set_title('Signal Composite (sin(t) + 0.5sin(2t) + 0.3sin(3t))')
axes[1, 1].set_xlabel('Temps (t)')
axes[1, 1].set_ylabel('Amplitude')
axes[1, 1].legend()
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
