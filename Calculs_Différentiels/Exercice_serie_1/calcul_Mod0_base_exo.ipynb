### Exercice 1.1 : Intervalles et Ensembles

**Objectif :** Comprendre les intervalles et leurs représentations

```python
def visualiser_intervalle(a, b, type_intervalle='ferme'):
    """
    Visualise un intervalle sur la droite réelle.

    Paramètres:
    -----------
    a, b : float
        Bornes de l'intervalle
    type_intervalle : str
        'ferme' : [a, b]
        'ouvert' : (a, b)
        'semi_ouvert_gauche' : (a, b]
        'semi_ouvert_droit' : [a, b)
    """
    # TODO: Créer une figure simple (axhline)
    fig, ax = plt.subplots(figsize=(10, 2))

    # TODO: Tracer la ligne représentant l'intervalle
    if type_intervalle == 'ferme':
        left_marker = 'filled'
        right_marker = 'filled'
        line_style = '-'
    elif type_intervalle == 'ouvert':
        left_marker = 'hollow'
        right_marker = 'hollow'
        line_style = '--'
    elif type_intervalle == 'semi_ouvert_gauche':
        left_marker = 'hollow'
        right_marker = 'filled'
        line_style = '-.'
    elif type_intervalle == 'semi_ouvert_droit':
        left_marker = 'filled'
        right_marker = 'hollow'
        line_style = '-.'
    else:
        raise ValueError(f"Type d'intervalle non reconnu: {type_intervalle}")

    # TODO: Ajouter les bornes (points pleins ou vides selon le type) [a, b], [a, b), (a,b), (a,b]
    ax.plot([a, b], [0, 0], line_style, color='red', linewidth=3)

    if left_marker == 'filled':
        ax.plot(a, 0, 'o', color='red', markersize=10)
    else:
        ax.plot(a, 0, 'o', markeredgecolor='red', markerfacecolor='white',
                markersize=10, markeredgewidth=2)

    if right_marker == 'filled':
        ax.plot(b, 0, 'o', color='red', markersize=10)
    else:
        ax.plot(b, 0, 'o', markeredgecolor='red', markerfacecolor='white',
                markersize=10, markeredgewidth=2)

    ax.text(a, -0.05, f'a = {a}', ha='center', va='top', fontsize=12)
    ax.text(b, -0.05, f'b = {b}', ha='center', va='top', fontsize=12)

    margin = (b - a) * 0.2
    ax.set_xlim(a - margin, b + margin)
    ax.set_ylim(-0.2, 0.2)

    ax.set_yticks([])

    titre = f"Intervalle {type_intervalle} : "
    if type_intervalle == 'ferme':
        titre += f"[{a}, {b}]"
    elif type_intervalle == 'ouvert':
        titre += f"({a}, {b})"
    elif type_intervalle == 'semi_ouvert_gauche':
        titre += f"({a}, {b}]"
    elif type_intervalle == 'semi_ouvert_droit':
        titre += f"[{a}, {b})"

    ax.set_title(titre, fontsize=14)

    ax.grid(True, alpha=0.3, linestyle='--')

    plt.tight_layout()
    plt.show()

if __name__ == "__main__":
    # Test avec différents types d'intervalles
    print("Exemple 1: Intervalle fermé [1, 5]")
    visualiser_intervalle(1, 5, 'ferme')

    print("\nExemple 2: Intervalle ouvert (2, 7)")
    visualiser_intervalle(2, 7, 'ouvert')

    print("\nExemple 3: Intervalle semi-ouvert à gauche (3, 6]")
    visualiser_intervalle(3, 6, 'semi_ouvert_gauche')

    print("\nExemple 4: Intervalle semi-ouvert à droite [4, 8)")
    visualiser_intervalle(4, 8, 'semi_ouvert_droit')
    pass
```

```python
def resoudre_inequation_valeur_absolue(a, epsilon):
    """
    Résout |x - a| < epsilon

    Solution: a - epsilon < x < a + epsilon
    """
    # TODO: Calculer la borne inférieure
    borne_inf = a - epsilon
    # TODO: Calculer la borne supérieure
    borne_sup = a + epsilon

    print(f"Résolution de |x - {a}| < {epsilon}")
    print(f"L'ensemble solution est l'intervalle: ({borne_inf}, {borne_sup})")
    return borne_inf, borne_sup
```

### Exercice 3.1 : Visualisation de la Tangente

```python
def visualiser_tangente(func, derivee, x0, nom="f(x)"):
    """
    Visualise une fonction et sa tangente en un point.
    """
    # TODO: Calculer le point y0 = f(x0)
    y0 = func(x0)
    # TODO: Calculer la pente m = f'(x0)
    pente = derivee(x0)

    # TODO: Définir l'équation de la tangente: y = y0 + m(x - x0)
    tangente = lambda x: y0 + pente * (x - x0)

    # Plotting code (laissé partiel pour l'étudiant)
    x = np.linspace(x0 - 2, x0 + 2, 100)
    plt.plot(x, func(x), label=nom)
    # TODO: Tracer la tangente
    # plt.plot(x, tangente(x), ...)
    plt.plot(x, tangente(x), 'r--', label=f"Tangente en x={x0}", linewidth=2)
    plt.legend()
    plt.show()
```

```python
# Visualiser la tangente au point (1,2) pour la fonction f(x) = x^2 + sin(x)
a = 1
f = lambda x : x**2 + np.sin(x)
f_prime = lambda x : 2*x + np.cos(x)
visualiser_tangente(f, f_prime, a)
```
