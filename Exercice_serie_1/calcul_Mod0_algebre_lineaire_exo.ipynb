## 2. Produit Scalaire et Orthogonalité

```python
# Produit scalaire et angle entre vecteurs
def angle_between_vectors(u, v):
    """Calcule l'angle entre deux vecteurs en degrés"""
    # TODO: Calculer le produit scalaire u.v
    dot_product = np.dot(u, v)
    # TODO: Calculer les normes de u et v
    norm_u = np.linalg.norm(u)
    norm_v = np.linalg.norm(v)
    # TODO: Calculer le cosinus de l'angle
    cos_angle = dot_product / (norm_u * norm_v)
    # TODO: Convertir en degrés (utiliser arccos)
    angle_rad = np.arccos(cos_angle)
    return np.degrees(angle_rad)
# Test
u = np.array([1, 0, 1])
v = np.array([1, 1, 1])
print(f"Angle entre {u} et {v}: {angle_between_vectors(u, v):.2f}° (Attendu: 45.00°)")
```

## 5. Matrice Inverse

```python
# Calcul de l'inverse d'une matriceprint("INVERSION DE MATRICE\n" + "="*60 + "\n")
A = np.array([[2, 1, 1],   [1,0,1], [4, 5, 3]])
print("Matrice A:")
print(A)# Déterminant

# TODO: Calculer le déterminant de A
det_A = np.linalg.det(A)
print(f"\nDéterminant: det(A) = {det_A}")

if det_A != 0:
# TODO: Calculer l'inverse de A
  A_inv = np.linalg.inv(A)
  print("\nInverse A⁻¹:")
  print(A_inv)

# Vérification
print("\nVérification A · A⁻¹ (doit être l'identité):")
print(np.dot(A, A_inv))
```

### Vérification de la solution par substitution

```python
import numpy as np

# Définir la matrice des coefficients A
A = np.array([[2, 1, -1], [1, -3, 5], [3, -1, 7]])

# Définir le vecteur second membre b
b = np.array([5, -4, 1])

# Résoudre le système pour trouver x
x_sol = np.linalg.solve(A, b)

print(f"Solution trouvée: x={x_sol[0]:.4f}, y={x_sol[1]:.4f}, z={x_sol[2]:.4f}\n")

# Substitution dans chaque équation originale

equation1_check = 2 * x_sol[0] + x_sol[1] - x_sol[2]
equation2_check = x_sol[0] - 3 * x_sol[1] + 5 * x_sol[2]
equation3_check = 3 * x_sol[0] - x_sol[1] + 7 * x_sol[2]

print(f"Vérification Équation 1 (2x + y - z = 5): {equation1_check:.4f} (Attendu: 5)")
print(f"Vérification Équation 2 (x - 3y + 5z = -4): {equation2_check:.4f} (Attendu: -4)")
print(f"Vérification Équation 3 (3x - y + 7z = 1): {equation3_check:.4f} (Attendu: 1)")

# Vérifier si les résultats sont très proches des valeurs attendues
tol = 1e-9 # Tolérance pour la comparaison en virgule flottante

if abs(equation1_check - 5) < tol and \
   abs(equation2_check - (-4)) < tol and \
   abs(equation3_check - 1) < tol:
    print("\nLa solution est correcte et vérifie toutes les équations du système.")
else:
    print("\nLa solution ne vérifie pas exactement toutes les équations du système.")
```

## 7. Application ML: Régression Linéaire

```python
# Régression linéaire avec algèbre linéaire
print("RÉGRESSION LINÉAIRE PAR MOINDRES CARRÉS\n" + "="*60 + "\n")
# Données synthétiquesnp.random.seed(42)
n = 20
x_data = np.linspace(0, 10, n)
y_true = 2*x_data + 3
# Vraie relation
y_data = y_true + np.random.normal(0, 2, n)
# Avec bruit# Matrice de design X (colonne de 1 et colonne de x)

# TODO: Créer la matrice X (indice: utiliser np.vstack et .T)
X = np.vstack((np.ones(n), x_data)).T

# Résolution par équation normale:
# theta = (X.T X)^-1 X.T y

# TODO: Calculer theta
theta = np.linalg.inv(X.T @ X) @ X.T @ y_data
print(f"Paramètres estimés: Intercept={theta[0]:.2f}, Pente={theta[1]:.2f}")
print("Vrais paramètres: Intercept=3.00, Pente=2.00")
```
