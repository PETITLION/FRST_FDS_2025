## $Exercice\ 1\ :\ Permutation\ circulaire$

```python
# Déclaration et affectation des variables
a, b, c = 15, 23, 8
print("Avant permutation:", a, b, c)

# Permutation circulaire à droite: a ==> b ==> c ==> a
a, b, c = c, a, b
print("Après permutation :", a, b, c)
```

## $Exercice\ 2\ :\ Calcul\ de\ moyenne\ pondérée$

```python
import numpy as np

def moyenne_ponderee_numpy(valeurs, poids):
    """
    Version vectorisée avec numpy.
    """
    valeurs = np.array(valeurs)
    poids = np.array(poids)
    
    if valeurs.shape != poids.shape:
        raise ValueError("Les tableaux doivent avoir la même taille")
    
    somme_poids = np.sum(poids)
    if somme_poids == 0:
        raise ValueError("La somme des poids ne peut pas être zéro")
    
    return np.sum(valeurs * poids) / somme_poids


notes = np.array([83, 65, 76, 89])
coefficients = np.array([0.3, 0.2, 0.4, 0.1])
print(f"Moyenne pondérée (numpy): {moyenne_ponderee_numpy(notes, coefficients):.2f}")
```
