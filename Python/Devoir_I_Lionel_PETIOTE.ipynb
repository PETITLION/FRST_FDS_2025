## $Exercice\ 2\ :\ Calcul\ de\ moyenne\ pondérée$

```python
# Fonction de récupération
def demander_liste(message):
    while True:
        try:
            valeurs = input(message).strip().split()
            valeurs = [float(v) for v in valeurs]
            return valeurs
        except ValueError:
            print("Erreur : veuillez entrer uniquement des nombres séparés par des espaces.\n")

# Fonction de moyenne pondérée
def moyenne_ponderee(valeurs, poids):
    valeurs = np.array(valeurs)
    poids = np.array(poids)

    somme_poids = np.sum(poids)
    if somme_poids == 0:
        print("Erreur : la somme des coefficients ne peut pas être égale à zéro.")
        return None  # on renvoie None si erreur

    return np.sum(valeurs * poids) / somme_poids

# Récupération des séries
while True:
    notes = demander_liste("Entrer les notes séparées par des espaces : ")
    coeffs = demander_liste("Entrer les coefficients séparés par des espaces : ")

    if len(notes) != len(coeffs):
        print("Les deux listes doivent avoir la même dimension ! Recommencez.\n")
        continue

    if sum(coeffs) == 0:
        print("Erreur : la somme des coefficients ne peut pas être égale à zéro. Recommencez.\n")
        continue

    break

# Calcul
moyenne = moyenne_ponderee(notes, coeffs)
if moyenne is not None:
    print(f"La moyenne pondérée est de : {moyenne:.2f}")
```

## $Exercice\ 3\ :\ Analyseur\ de\ texte\ basique$

```python
phrase = input("Vueillez entrer une phrase : ")

mots = phrase.split()

nombre_de_mots = len(mots)

mot_le_plus_long = ""
if mots:
    for mot in mots:
        if len(mot) > len(mot_le_plus_long):
            mot_le_plus_long = mot

nombre_total_de_lettres = 0
if mots:
    for mot in mots:
        nombre_total_de_lettres += len(mot)
    moyenne_lettres_par_mot = nombre_total_de_lettres / nombre_de_mots
else:
    moyenne_lettres_par_mot = 0

print(f"\nVoici votre phrase : '{phrase}'")
print(f"- Nombre de mots : {nombre_de_mots}")
print(f"- Mot le plus long : '{mot_le_plus_long}'")
print(f"- Nombre moyen de lettres par mot : {moyenne_lettres_par_mot:.2f}")
```

## $Exercice\ 5\ :\ Vérificateur\ de\ palindrome$

```python
mot = input("Entrer un mot : ")

mot_inverse = mot[::-1]

if mot.upper() == mot_inverse.upper():
    print(f"Le mot '{mot}' est un palindrome.")

else:
    print(f"Le mot '{mot.upper()}' n'est pas un palindrome.")
```

## $Exercice\ 9\ :\ Codeur/décodeur\ simple$

```python
alphabet = "abcdefghijklmnopqrstuvwxyz"
choix = input("Voulez-vous encoder(1) ou décoder(2) un message ? : ")

if choix == "1":
    message = input("Entrer le message à encoder : ")
    message_code = ""
    for lettre in message.lower():
        if lettre in alphabet:
            index_original = alphabet.find(lettre)
            index_code = (index_original + 3) % len(alphabet)
            message_code += alphabet[index_code]
        else:
            # Conserver les caractères qui ne sont pas dans l'alphabet (espaces, ponctuation, etc.)
            message_code += lettre

    print(f"Le message codé est : '{message_code}'")

elif choix == "2":
    message = input("Entrer le message à décoder : ")
    message_decode = "" # Initialisation de message_decode
    for lettre in message.lower():
        if lettre in alphabet:
            index_original = alphabet.find(lettre)
            index_code = (index_original - 3) % len(alphabet)
            message_decode += alphabet[index_code] # Correction de l'opérateur
        else:
            # Conserver les caractères qui ne sont pas dans l'alphabet (espaces, ponctuation, etc.)
            message_decode += lettre # Correction de la variable
    print(f"Le message décodé est : '{message_decode}'")
else:
    print("Choix invalide. Veuillez entrer '1' pour encoder ou '2' pour décoder.")
```

## $Exercice\ 10\ :\ Calculateur\ de\ pourboire$

```python
while True:
    montant_consomme_str = input("Entrer le montant consommé : ")
    try:
        montant_consomme = float(montant_consomme_str)
        break
    except ValueError:
        print("Erreur : le montant consommé doit être un nombre. Veuillez réessayer.")

while True:
    pourboire_str = input("Entrer le pourcentage de pourboire : ")
    try:
        pourboire = float(pourboire_str)
        break
    except ValueError:
        print("Erreur : le pourcentage de pourboire doit être un nombre. Veuillez réessayer.")

montant_pourboire = (pourboire / 100) * montant_consomme
print(f"\n Le montant de pourboire est de {montant_pourboire:.2f} gourdes.")
montant_total = montant_consomme + montant_pourboire
print(f"\n Le montant total à payer est de {montant_total:.2f} gourdes.")
```
